```python
import torch
import numpy as np
from tqdm import tqdm
```

```python
num_timesteps = 20
```

```python
x_shape=(5, 1, 28, 28)

x = torch.randn(x_shape,)

batch_size = x_shape[0]
x = torch.randn(x_shape,)

for i in range(num_timesteps, 0, -1):
    timestep = torch.tensor(
        [i] * batch_size,
        dtype=torch.long,
    )
    display(timestep)
```

```python
T = 500
ddim_timestep = 50

# DDIMでの1ステップをDDPMのステップに変換
ddim_timesteps = torch.linspace(1, T, ddim_timestep, dtype=torch.long)

for i in tqdm(range(ddim_timestep, 0, -1)):
    timestep = torch.tensor(
        [ddim_timesteps[i - 1]] * batch_size, dtype=torch.long,
    )
    if i == 1:
        timestep_prev = torch.tensor(
            [0] * batch_size,
            dtype=torch.long,
        )
    else:
        timestep_prev = torch.tensor(
            [ddim_timesteps[i - 2]] * batch_size,
            dtype=torch.long,
        )
    print(f"i: {i}")
    print(f"times_step: {timestep[0]}")
    print(f"times_step_prev: {timestep_prev[0]}")
```
